```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
product_name = []
description_name = []
price = []
```

```python
    url = 'https://www.flipkart.com/search?q=mobile+under+50000&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_18_na_na_ps&otracker1=AS_QueryStore_OrganicAutoSuggest_1_18_na_na_ps&as-pos=1&as-type=RECENT&suggestionId=mobile+under+50000%7CMobiles&requestId=604b1ef6-f53a-4a4e-9f0f-f91d5befdcbb&as-searchtext=mobile+under+50000&page=1'
    r = requests.get(url)
    soup = BeautifulSoup(r.text,'lxml')
    np = soup.find('a',class_ = 'ge-49M _2Kfbh8').get('href')
    cnp = 'https://www.flipkart.com'+np
    print(cnp)
```

```python
names = soup.find_all('div',class_ = '_4rR01T')
print(names)
for i in names:
    names = i.text
    product_name.append(names)
print(product_name) 
```

```python
print(len(product_name))
```

```python
description = soup.find_all('div',class_ = 'fMghEO')
for i in description:
    description = i.text
    description_name.append(description)
print(description_name)
```

```python
print(len(description_name))
```

```python
rate = soup.find_all('div',class_ = '_30jeq3 _1_WHN1')
for i in rate:
    rate= i.text
    price.append(rate)
print(price)
```

```python
print(len(price))
```

```python
df = pd.DataFrame({'product_name':product_name,'description_name':description_name,'price':price})
```

```python
df.to_csv('F:/independent projects/mobile_under_50000')
```
